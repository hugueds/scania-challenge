```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open('./islands.txt', 'r')

text = file.read()

arr = np.zeros((1000,1000, 3))

column = 0
line = 0

for i in text:
    if column < 1000 and i != '\n':
        arr[line][column] = int(i)
        column += 1
    else:
        column = 0
        line += 1

islands = 0
```

```python
print(arr.shape)

plt.figure(figsize=(20,20))
plt.imshow(arr, cmap='gray')
```

```python
small = arr.copy()
small = arr[:10,:10]
color = 0

i = 0
j = 0
c = 0

islands = 0
new_island = false
is_island = false

for i in range(small.shape[0]):
    for j in range(small.shape[1]):
        if small[i][j].any() > 0:
            islands += 1

print(islands)
    

plt.imshow(small, cmap='gray')
```
