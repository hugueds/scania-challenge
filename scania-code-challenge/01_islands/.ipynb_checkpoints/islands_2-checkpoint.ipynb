WRITE A PROGRAM THAT COUNTS ALL INDIVIDUAL ISLANDS IN THIS FILE

### PROBLEM
The file contains a grid of 1000 x 1000 cells filled with 1 and 0 values. Your task is to count the number groups with connecting 1.

To count as a connection the values must be directly connection up, down, left or right. Hence diagonal is not counted as a connection.

In the example to the left there are 6 islands of 1's.

### ANSWER
The answer is the total number of islands in the file.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
file = open('./islands.txt', 'r')
text = file.read()


np_arr = np.zeros((1000,1000))
arr = []
sub_arr = []

column = 0
line = 0

for i in text:
    if column < 1000 and i != '\n':
        np_arr[line][column] = int(i)       
        sub_arr.append(i)
        column += 1
    else:        
        column = 0
        line += 1
        arr.append(sub_arr)
        sub_arr = []

print(len(arr))
print(f'SHAPE {np_arr.shape}')

plt.figure(figsize=(10,10))
plt.imshow(np_arr, cmap='gray')
```

```python
class Solution(object):
    def numIslands(self, grid):
        if not grid: return 0
        r, c = len(grid), len(grid[0])
        visited = [[False for _ in range(c)] for _ in range(r)]

        def dfs(i, j):
            if i < 0 or i >= r or j < 0 or j >= c or grid[i][j] == 0 or visited[i][j]:
                return
            visited[i][j] = True
            dfs(i + 1, j)
            dfs(i - 1, j)
            dfs(i, j + 1)
            dfs(i, j - 1)

        count = 0
        for i in range(r):
            for j in range(c):
                if not visited[i][j] and grid[i][j] == 1:
                    dfs(i, j)
                    count += 1
        return count
```
